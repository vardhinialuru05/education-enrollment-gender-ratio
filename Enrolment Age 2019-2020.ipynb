```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('enrolment_age_2019_20.csv')
df.head()
```

```python
df.drop(['ac_year'], axis = 1, inplace = True)
df.head()
```

```python
cols = list(df.columns)
cols
```

```python
df['boys_total'] = df[cols[5:17]].sum(axis=1)
df.head()
```

```python
df['girls_total'] = df[cols[17:29]].sum(axis=1)
df.head()
```

```python
cols = list(df.columns)
cols
```

```python
# df_visual_1 = df[cols[1, 2, 3, 4, -1, -2]]
df_visual_1 = df[cols[2:3] + [cols[-1], cols[-2]]]

df_visual_1.head()
```

```python
temp_list = df_visual_1.values.tolist()
temp_list
```

```python
new_dictionary = {}
for row in temp_list:
    if new_dictionary.get(row[0]) is None:
        new_dictionary[row[0]] = [row[1], row[2]]
    else:
        new_dictionary[row[0]] = list(np.add(new_dictionary[row[0]], [row[1], row[2]]))

new_dictionary
```

```python
for row in new_dictionary:
    new_dictionary[row] = (new_dictionary[row])[0] / (new_dictionary[row])[1]
new_dictionary
```

```python
X = np.array(list(new_dictionary.keys()))
Y = np.array(list(new_dictionary.values()))


col = []
for val in Y:
    if val > 1:
        col.append('orchid')
    else:
        col.append('royalblue')




plt.rcParams["figure.figsize"] = [15, 7]
plt.bar(X, Y, color = col)
plt.xlabel('States', fontweight='bold', color = 'black', fontsize='17', horizontalalignment='center', labelpad=15)
plt.ylabel('Girls to Boys Ratio', fontweight='bold', color = 'black', fontsize='17', verticalalignment='center', labelpad=30)
plt.xticks(rotation = 90)
plt.axhline(1, color='red', linestyle='--')

plt.show()
```
